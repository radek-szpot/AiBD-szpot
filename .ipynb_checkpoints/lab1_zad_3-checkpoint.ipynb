Ćwiczenie przedstawiające wykres zadanej funkcji na zadanym przedziale.

Import bibliotek oraz zrealizowanie funkcji:   
$f(x)=x^2+5$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2+5
```

Wygenerowanie przedziałów:   
1. $x>-1$ oraz $x<1$   
2. $x>-6$ oraz $x<6$   
3. $x>0$ oraz $x<5$   

wykorzystując funkcję linspace z biblioteki numpy.

```python
przedzial1 = np.linspace(-1, 1)
przedzial2 = np.linspace(-6, 6)
przedzial3 = np.linspace(0, 5)
```

Wizualizacja wykresu funkcji na przedziale pierwszym:

```python
plt.plot(przedzial1,f(przedzial1))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("wykres funkcji $x^2+5$ na przedziale pierwszym")
plt.legend(["f(x)"])
plt.show
```

Wizualizacja wykresu funkcji na przedziale drugim:

```python
plt.plot(przedzial2,f(przedzial2))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("wykres funkcji $x^2+5$ na przedziale drugim")
plt.legend(["f(x)"])
plt.show
```

Wizualizacja wykresu funkcji na przedziale trzecim:

```python
plt.plot(przedzial3,f(przedzial3))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("wykres funkcji $x^2+5$ na przedziale trzecim")
plt.legend(["f(x)"])
plt.show
```
