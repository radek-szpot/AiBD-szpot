```python
import numpy as np
import pandas as pd
```

```python
with open('../Original-data/weather.txt') as f:
    lines = f.readlines()
    
# stworz tabele na dane z 2010 roku
required_data = []
    
for line in lines:
    # usun spacje zrob tablice samych elementow
    line_without_space = line.split()
    if len(line_without_space) > 1 and line_without_space[0][11:15] == "2010":
        for elem in line_without_space:
            # usun elementy "I"
            if elem == "I":
                line_without_space.remove("I")
                
        # polacz falgi S z elementami        
        index = 0
        for elem in line_without_space:
            if elem == "S":
                if index == len(line_without_space)-1:
                    line_without_space[1] = "S" + line_without_space[1]
                    del line_without_space[-1]
                else:
                    line_without_space[index+1] = "S" + line_without_space[index+1]
                    del line_without_space[index]
            index += 1
        
        # jesli za duzo badz za malo danych - dane uszkodzone
        if len(line_without_space) > 1:
            # sprawdz czy nie ma skeljonej daty w 1 elemncie   
            if len(line_without_space[0]) > 21:
                    first_el, sec_el = line_without_space[0][0:21], line_without_space[0][21::]
                    line_without_space[0] = first_el
                    line_without_space.insert(1, sec_el)
            if len(line_without_space) == 32:
                required_data.append(line_without_space)
 

            
# czyszenie danych (element -9999 jest błędny)
for line in required_data:
    for index, measure in enumerate(line[1::]):
        if measure == "S-9999" or measure == "-9999":
            line[index+1] = np.nan

# podzial pierwszego elementu na id, rok, miesiąc, nazwa zmiennych
for line in required_data:
    merged = line[0] 
    del line[0]
    line.insert(0, merged[17:21])
    line.insert(0, merged[15:17])
    line.insert(0, merged[11:15])
    line.insert(0, merged[0:11])
```

```python
weather = pd.DataFrame(required_data, columns=["ID", "YEAR", "MONTH", "TYPE"] + [i+1 for i in range(31)])
weather
```

```python
# podzial danych tak aby kazda zmienna tworzyla kolumne, a kazda obserwacja tworzyla rząd

# stworz listy na dane do DataFrame
id_lst = []
year_lst = []
month_lst = []
type_lst = []
day_lst = []
value_lst = []
flag_lst = []

# w petli dodaj dane tak aby kazda obserwacja miala wszystkie potrzebne informacje
for line in required_data:
    id_lst = id_lst + [line[0]]*31
    year_lst = year_lst + [line[1]]*31
    month_lst = month_lst + [line[2]]*31
    type_lst = type_lst + [line[3]]*31
    day_lst = day_lst + [i+1 for i in range(31)]
    # jesli obserwacja ma flage to nalezy to zaznaczyc i usunac falge z wartosci    
    for elem in line[4::]:
        type_s = False
        if type(elem) == str:
            for char in elem:
                if char == "S":
                    flag_s = True
                    break
                else:
                    flag_s = False
            if flag_s:
                flag_lst = flag_lst + ["S"]
                value_lst = value_lst + [elem[1::]]
            else:
                flag_lst = flag_lst + ["   "]
                value_lst = value_lst + [elem]
        else:
            flag_lst = flag_lst + ["   "]
            value_lst = value_lst + [elem]
```

```python
# uwtorz DataFrame z wygenerowanych list
split_data = {"ID" : id_lst, "YEAR" : year_lst, "MONTH" : month_lst, "DAY" : day_lst, "DATA TYPE": type_lst, "VALUE": value_lst, "FLAG": flag_lst}
split_data_df = pd.DataFrame(split_data)
split_data_df
split_data_cleared.to_csv(path_or_buf="../Analisys-data/all_data_data_with_nan")
```
