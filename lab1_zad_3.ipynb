Ćwiczenie przedstawiające wykres zadanej funkcji na zadanym przedziale

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2+5

przedzial1 = np.linspace(-1, 1)
przedzial2 = np.linspace(-6, 6)
przedzial3 = np.linspace(0, 5)
```

```python
plt.plot(przedzial1,f(przedzial1))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("wykres funkcji x^2+5")
plt.legend(["f(x)"])
plt.show
```

```python
plt.plot(przedzial2,f(przedzial2))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("wykres funkcji x^2+5")
plt.legend(["f(x)"])
plt.show
```

```python
plt.plot(przedzial3,f(przedzial3))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("wykres funkcji x^2+5")
plt.legend(["f(x)"])
plt.show
```
